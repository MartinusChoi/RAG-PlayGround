```python
import os
from glob import glob
from typing import Callable, List
from dotenv import load_dotenv
import nest_asyncio
import pandas as pd

from llama_index.core import Document
from kiwipiepy import Kiwi

from llama_index.core.node_parser import SemanticSplitterNodeParser
from llama_index.embeddings.openai import OpenAIEmbedding
```

```python
parsed_path = os.path.join(root_dir, "..", "Parsing", "hybrid_parse", "parsed_result.parquet")
result = pd.read_parquet(parsed_path)
result
```

```python
texts = result["texts"].tolist()
texts
```

```python
documents = [Document(text=texts[0])]
```

```python
def split_by_sentence_kiwi() -> Callable[[str], List[str]]:
    kiwi = Kiwi()

    def split(text: str) -> List[str]:
        kiwi_result = kiwi.split_into_sents(text)
        sentences = list(map(lambda x: x.text, kiwi_result))

        return sentences
    
    return split
```

```python
embed_model = OpenAIEmbedding()

semantic = SemanticSplitterNodeParser(
    sentence_splitter=split_by_sentence_kiwi(), 
    embed_model=embed_model,
    breakpoint_percentile_threshold=95
)
semantic_result = semantic.get_nodes_from_documents(documents=documents)
semantic_texts = [x.text for x in semantic_result]
```
