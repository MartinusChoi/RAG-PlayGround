# Table Hybrid Parse with AutoRAG

## 0. Import Library & Set Environment Variables

```python
import pandas as pd
from glob import glob
import os
from dotenv import load_dotenv
import nest_asyncio

from autorag.parser import Parser
```

```python
load_dotenv()
```

```python
nest_asyncio.apply()
```

```python
root_dir = os.getcwd()
data_dir = os.path.join(root_dir, "data", "parse_single")
glob_path = os.path.join(data_dir, "*")
```

## 1. Parse with AutoRAG

```python
pjt_dir = os.path.join(root_dir, "hybrid_parse")

parser = Parser(data_path_glob=glob_path, project_dir=pjt_dir)
```

```python
yaml_path = os.path.join(root_dir, "config", "hybrid_parse.yaml")

parser.start_parsing(yaml_path=yaml_path)
```

## 2. Check Result

```python
result = pd.read_parquet(os.path.join(pjt_dir, "parsed_result.parquet"))

result
```

```python
print(result.loc[result["page"]==5, 'texts'].tolist()[0])
```
