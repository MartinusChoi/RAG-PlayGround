## 1. Data Cleaning

> PDF 파일 자체에 존재하는 불필요한 정보나 페이지를 제거하는 과정

> 페이지 마다 반복되는 불필요한 정보를 제거하는 방법도 좋으나, 일일히 직접 제거해야 하기에 효율이 떨어짐

> 필요하지 않은 페이지(표지 등)도 직접 제거하는 불편함이 있으나, 훨씬 수고도 덜하며 효과도 좋다

## 4. Chunking

- 크게 Token Text Splitter, Character Spliter, Sentence Splitter, Sentence Window Splitter, Semantic Chunking 5가지 옵션이 존재한다
    - 이 중, 통상적으로 Token SPlitter, Window, Semantic이 성능이 좋다

### 4-1. Basic Splitter

```python
parsed_path = os.path.join(root_dir, "autorag_project", "multimodal", "parsed_result.parquet")
pjt_dir = os.path.join(root_dir, "autorag_project", "basic_chunking")
yaml_path = os.path.join(root_dir, "config", "chunking", "basic_chunking.yaml")
```

```python
chunker = Chunker.from_parquet(
    parsed_data_path=parsed_path,
    project_dir=pjt_dir
)
```

```python
chunker.start_chunking(
    yaml_path=yaml_path
)
```

### 4-2. Semantic Splitter

```python
pjt_dir = os.path.join(root_dir, "autorag_project", "semantic_chunking")
yaml_path = os.path.join(root_dir, "config", "chunking", "semantic_chunking.yaml")
```

```python
chunker = Chunker.from_parquet(
    parsed_data_path = parsed_path,
    project_dir = pjt_dir
)
```

```python
chunker.start_chunking(
    yaml_path=yaml_path
)
```
